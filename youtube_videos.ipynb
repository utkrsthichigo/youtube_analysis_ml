```python
# Define a function to hash the video_id value
def hash_video_id(video_id):
    return hash(video_id)

# Apply the hash_video_id function to the video_id column
df['hashed_video_id'] = df['video_id'].apply(hash_video_id)

```

```python
from sklearn.decomposition import PCA

# Create an instance of the PCA class
#pca = PCA(n_components=2)


# Define features to include in PCA
features = ['duration_video','view_count_video','likes_video','Total_Comments','Total_Subscriber','view_count_channel']

# Standardize features
df[features] = (df[features] - df[features].mean()) / df[features].std()

# Fit PCA
pca = PCA()
pca.fit(df[features])

# Transform data using PCA components
df_pca = pd.DataFrame(pca.transform(df[features]), columns=[f'component_{i+1}' for i in range(len(features))])

# Create a composite growth score based on PCA components
df['growth_score'] = df_pca.sum(axis=1)

# Save data with growth score column
#df.to_csv('channel_data_with_growth_score.csv', index=False)
```
